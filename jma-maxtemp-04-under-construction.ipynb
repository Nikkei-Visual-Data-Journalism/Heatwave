```python
import pandas as pd
```

```python
#観測地点の一覧
url = 'https://raw.githubusercontent.com/Nikkei-Visual-Data-Journalism/Heatwave/main/data-maxtemp/meta/points_list.csv'
points = pd.read_csv(url)
```

```python
#県別の気温（真夏日・猛暑日）、過去分のデータを呼び出し
url = "./data-maxtemp/timeseries-data/jma-maxtemp-temp-by-pref-ts.csv"
by_pref = pd.read_csv(url)

#確認
print(by_pref.pref.nunique())

by_pref.head()
```

```python
#データ漏れのチェック
test = by_pref.pivot(index='date',columns='pref',values='maxtemp_capitol')
test = test.iloc[:-3].fillna('no_data')
no_data = test.where(test == 'no_data').dropna(how='all').dropna(how='all',axis=1)
```

```python
no_data_pref = test.where(test == 'no_data').dropna(how='all').dropna(how='all',axis=1).columns.to_list()
```
