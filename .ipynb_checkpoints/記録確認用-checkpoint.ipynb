```python
import pandas as pd
```

```python
url = './data-maxtemp/timeseries-data/jma-maxtemp-heatpoints-by-pref-daily.csv'
```

```python
df = pd.read_csv(url, parse_dates=['date'])
```

東京都の記録

```python
tokyo = df.loc[df.pref=='東京都'].copy()
```

```python
#真夏日の連続日数
def count_consecutive_true(group):
    is_true = group['over30_capitol']
    shifted = is_true.shift(fill_value=False)
    reset_points = ~is_true | (~shifted & is_true)
    group['consecutive_count'] = reset_points.cumsum().where(is_true, 0)
    group['consecutive_count'] = group.groupby('consecutive_count').cumcount() + 1
    return group
```

```python
tokyo = tokyo.groupby('year').apply(count_consecutive_true).reset_index(drop=True)
```
